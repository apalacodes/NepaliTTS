```python
# Audio Preprocessing and Normalization
# 22.05 kHz, and Stero = mono
# For the OpenSLR43 - may not need any audio filters as the clips barely contain any silent frames
```

```python
import os

input_dir = '/content/drive/MyDrive/TTSMajorProject/Notebooks/data/openslr/SLR43/ne_np_female/wavs'
output_dir = '/content/drive/MyDrive/TTSMajorProject/Notebooks/data/pre_processed'
os.makedirs(output_dir, exist_ok=True)
```

```python
import subprocess

def preprocess_ffmpeg(input_file, output_file):

    # the FFmpeg command
    cmd = [
        "ffmpeg",
        "-i", input_file,
        "-ar", "22050",        # resample to 22.05 kHz
        "-ac", "1",           # Convert to mono
        # "-af", ",".join(audio_filters),# apply audio filters
        "-f", "wav", # Preserve format
        "-y",
        output_file                   # overwrite
    ]

    subprocess.call(cmd)

```

```python
for filename in os.listdir(input_dir):
    if filename.endswith('.wav'):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)
        preprocess_ffmpeg(input_path, output_path)
        print(f"Processed: {filename}")
```
